```python
                 # CONVERT SOME STRING VALUE TO FLOAT


def clean_and_convert(value):
    if isinstance(value, str):
        # Replace '18_' with '18' and attempt conversion to float
        cleaned_value = value.replace('_', '')
        try:
            return float(cleaned_value)
        except ValueError:
            return value  # Return original value if conversion fails
    return value  # Return value unchanged if it's not a string

# Apply the function to each cell in the DataFrame
data = data.applymap(clean_and_convert)
```

                                                         BUILDING THE MODEL

```python
from sklearn.model_selection import train_test_split
x = np.array(data[['Annual_Income','Num_Bank_Accounts','Num_Credit_Card','Num_of_Loan','Delay_from_due_date','Num_of_Delayed_Payment',
                  'Outstanding_Debt','Monthly_Balance']])
y = np.array(data[['Credit_Mix']])
```

```python
data['Credit_Mix'].shape
```

```python
print(x.shape) 
print(y.shape)  
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
  # this is for the NaN value in a y_test
non_nan_values = y_test[~np.isnan(y_test)]    
unique, counts = np.unique(non_nan_values, return_counts=True)
most_frequent_value = unique[np.argmax(counts)]

# Fill NaN values with the most frequent value
y_test= np.where(np.isnan(y_test), most_frequent_value, y_test)

```

```python
from sklearn.ensemble import RandomForestClassifier
Model = RandomForestClassifier(criterion='entropy',n_estimators=150)
```

```python
Model.fit(x_train,y_train)
```

```python
y_predict=Model.predict(x_test)
```

                                                 ACCURACY OF THE MODEL

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
```
